# Exploratory Data Analysis-  exploring employment for college graduates from different majors

The dataset contains information aggreagted on recent college grads from different majors, where each data point represents a major

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
recent_grads= pd.read_csv("recent-grads.csv")
#Examining data
print(recent_grads.iloc[0])
print(recent_grads.head)
print(recent_grads.tail)
recent_grads.describe()
```

```python
#Finding number of rows
raw_data_count= recent_grads.shape[0]
recent_grads= recent_grads.dropna()
cleaned_data_count= recent_grads.shape[0]
print(cleaned_data_count)
```

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```

Looking at how the number of people of employed vs the number of people in each major's sample size. Strong correlation observed as is expected

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs. Sample_size')
```

No crrelation observed in the medium income of each major vs the sample size for each major

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment_rate vs. Sample_size')
```

No crrelation observed in the unemployment of each major vs the sample size for each major

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Median vs. Full_time')
```

No crrelation observed in the medium income of each major vs the the number of full time employees from each major

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Median vs. ShareWomen')
```

There seems to be a weak negative correlation between the share of women graduates in a major and the medium income

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
```

No correlation observed in the medium income vs the number male graduates from the major

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

No correlation observed in the medium income vs the number of female graduates from the major

Questions to ponder on:

1) Do students in more popular majors make more money?
No. It seems that the highest median salaries are for majors that have small sample sizes

2) Do students that majored in subjects that were majority female make more money?
It is the contrary. The plot for median income vs share of women seems to suggest that there is a negative correlation between those 2 factors

3) Is there any link between the number of full-time employees and median salary?
The highest median salaries seem to have a very low number of full-time employees. For the lower median salaries there are a wide range of full-time employees.
