```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(18, 16))

for sp in range(0,18):
    ax = fig.add_subplot(6,3,sp+1)
    
    # STEM column code
    if (sp+1)%3==1:
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp//3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp//3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(stem_cats[sp//3])
        
    # liberal arts column code    
    if (sp+1)%3==2:
        if sp//3==5:
            ax.spines["right"].set_visible(False)    
            ax.spines["left"].set_visible(False)
            ax.spines["top"].set_visible(False)    
            ax.spines["bottom"].set_visible(False)
            ax.set_yticklabels([])
            ax.set_xticklabels([])
            continue
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp//3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp//3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(lib_arts_cats[sp//3])
    # Other majors code    
    if (sp+1)%3==0:
        ax.plot(women_degrees['Year'], women_degrees[other_cats[sp//3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp//3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(other_cats[sp//3])
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.6)    
    
    if sp//3 == 0:
        ax.text(2005, 90, 'Women')
        ax.text(2005, 8, 'Men')
    elif (sp//3 == 5) |(sp==13):
        ax.text(2005, 65, 'Men')
        ax.text(2005, 30, 'Women')
        ax.tick_params(labelbottom='on')
plt.savefig('gender_degrees.png')
plt.show()
```
